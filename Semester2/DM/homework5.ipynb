```python
import pandas as pd
from apyori import apriori
from fpgrowth_py import fpgrowth
from time import time
```

```python
df = pd.read_csv('./datasets/lab5/movies.csv')
genres = df.genres.str.split('|')
```

## Apriori

### Extraction

```python
# support was reduced from 0.01 to 0.001 and confidence was reduced from 0.9 to 0.8, because there were no rules with initial parameters
apriori_start_time = time()
apriori_rules = list(apriori(
    genres,
    min_support=0.001,
    min_confidence=0.8,
    min_lift=1))
apriori_end_time = time()
```

### Execution Time

```python
print(f'Execution time: {apriori_end_time - apriori_start_time} seconds')
```

### Rules

```python
print(f'Apriori extracted {len(apriori_rules)} rules')

apriori_rules_df = pd.DataFrame(
    [
        {
            'Antecedent': rule[2][0][0],
            'Consequent': rule[2][0][1],
            'Support': rule[1],
            'Confidence': rule[2][0][2],
            'Lift': rule[2][0][3]
        } for rule in apriori_rules
    ]
)
apriori_rules_df
```

## FPGrowth

### Extraction

```python
fpgrowth_start_time = time()
freq_item_set, fpgrowth_rules = fpgrowth(genres, minSupRatio=0.001, minConf=0.8)
fpgrowth_end_time = time()
```

### Execution Time

```python
print(f'Execution time: {fpgrowth_end_time - fpgrowth_start_time} seconds')
```

### Rules

```python
print(f'FPGrowth extracted {len(fpgrowth_rules)} rules')

fpgrowth_rules_df = pd.DataFrame(
    [
        {
            'Antecedent': list(rule[0]),
            'Consequent': list(rule[1]),
            'Confidence': rule[-1],
      } for rule in fpgrowth_rules
    ]
)

fpgrowth_rules_df
```
