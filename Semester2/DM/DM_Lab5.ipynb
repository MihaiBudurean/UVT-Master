### Example 1:  Package apyori - https://pypi.org/project/apyori/

#### Remarks - apyori / Observații - apyori:
 * each RelationRecord corresponds to a frequent itemset / structura RelationRecord corespunde unui set de entități
 * each OrderedStatistics correspond to a rule (the antecedent is specified by items_base, the consequent is specified by items_add)/ fiecare componenta de tip OrderedStatistic corespunde unei reguli (antecedent: items_base, concluzie: items_add)

```python
#!pip install apyori
import apyori
from apyori import apriori
```

```python
# first example - small set of transactions
transactions = [['milk', 'bread', 'meat', 'water'],
['bread', 'water'],
['bread', 'butter', 'meat', 'water'],
['water'],
]
results = list(apriori(transactions))
```

```python
results
```

```python
# number of extracted rules / numar de reguli extrase
len(results)
```

#### Example 2 - extract association rules from the groceries list of transactions / Exemplu 2 - extragere reguli de asociere din fisier groceries.csv
Hint:  https://www.kaggle.com/code/raibabar/market-basket-analysis-with-apriori-algorithim

```python
transactions_groceries = []

with open('./datasets/lab5/groceries.csv') as f:
    for line in f:
        transaction = [item for item in line.strip().split(',') if item != 'NaN']
        transactions_groceries.append(transaction)
```

```python
len(transactions_groceries)
```

```python
# extragere reguli folosind algoritmul apriori
rules_groceries = list(apriori(
    transactions_groceries,
    min_support=0.001,
    min_confidence=0.9,
    min_length=2,
    max_length=5, min_lift=1))
```

```python
len(rules_groceries)
```

```python
# construct a data frame containing the rules / construirea unui tabel cu regulile
import pandas as pd

rules_df = pd.DataFrame(
    [{'Antecedent': list(rule[0])[0],
    'Consequent': list(rule[0])[1],
    'Support': rule[1],
    'Confidence': rule[2][0][2],
    'Lift': rule[2][0][3]} for rule in rules_groceries])

# sort the rules decreasingly by lift / sortare descrescatoare dupa lift
rules_df = rules_df.sort_values('Lift', ascending=False)
```

```python
rules_df.head(39)
```

#### Ex. 2.1. Analyze the influence of the thresholds for support (values: 0.001, 0.005, 0.01, 0.05) and confidence (values: 0.1, 0.3, 0.5, 0.7, 0.9).  / Analizati influenta pragurilorpentru suport (valori: 0.001, 0.005, 0.01, 0.05) si nivel de incredere (valori: 0.1, 0.3, 0.5, 0.7, 0.9).

```python
supp = [0.001, 0.005, 0.01, 0.05]
conf = [0.1, 0.3, 0.5, 0.7, 0.9]

for s in supp:
    for c in conf:
        rules_groceries = list(apriori(
            transactions_groceries,
            min_support=s,
            min_confidence=c,
            min_length=2,
            max_length=5,
            min_lift=1
        ))
        print(f'With support {s} and confidence {c} a total of {len(rules_groceries)} rules')
```

#### Ex. 2.2. Analyze the influence of the lift value (values: 1.5, 2, 3, 4) and print 10 rules in decreasing order of the lift / Analizati influenta valorii lift-ului (valori: 1.5, 2, 3, 4) si afisati primele 10 reguli in ordinea descrescatoare a valorii lift-ului.

```python
lifts = [1.5, 2, 3, 4]

for l in lifts:
    rules_groceries = list(apriori(
        transactions_groceries,
        min_support=0.005,
        min_confidence=0.5,
        min_length=2,
        max_length=5, min_lift=l))
    
    if len(rules_groceries):
        rules_df = pd.DataFrame(
            [
                {
                    'Antecedent': rule[2][0][0],
                    'Consequent': rule[2][0][1],
                    'Support': rule[1],
                    'Confidence': rule[2][0][2],
                    'Lift': rule[2][0][3]
                } for rule in rules_groceries
            ]
        )

        sorted_rules_df = rules_df.sort_values(by='Lift', ascending=False)

        if len(sorted_rules_df) > 10:
            print(sorted_rules_df.head(10))
        else:
            print(sorted_rules_df.head(len(sorted_rules_df)))
```

### Example 4. Package mlxtend (Machine Learning Library Extensions) - apriori algorithm
Source: https://github.com/jsrpy/Association_Rule/blob/master/association_rules_introduction.ipynb

```python
# !pip install mlxtend
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# data in one-hot encoded format
data = {'ID':[1,2,3,4,5,6],
       'Onion':[1,0,0,1,1,1],
       'Potato':[1,1,0,1,1,1],
       'Burger':[1,1,0,0,1,1],
       'Milk':[0,1,1,1,0,1],
       'Beer':[0,0,1,0,1,0]}
df = pd.DataFrame(data)
df
```

```python
frequent_itemsets = apriori(df[['Onion', 'Potato', 'Burger', 'Milk', 'Beer' ]], min_support=0.50, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules
```

```python
# select rules based on confidence and lift
rules [ (rules['lift'] >1.125)  & (rules['confidence']> 0.8)  ]
```
