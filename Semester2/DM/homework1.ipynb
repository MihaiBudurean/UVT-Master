```python
import pandas as pd
import numpy as np

hotel_reservations = pd.read_csv('./datasets/homework1-2/hotel_reservations.csv')

# changed Booking_ID from INN12345 to 12345, to be considered numerical
hotel_reservations['Booking_ID'] = hotel_reservations['Booking_ID'].apply(lambda x: np.int64(x[3:]))
```

b) Describe the dataset with respect to
- The number and types of attributes
- Distribution of the instances per classes

```python
# The number and types of attributes
print(f'Number of attributes: {hotel_reservations.shape[1]}')
print('Attributes of the dataset:')
print(hotel_reservations.dtypes)
```

```python
# Distribution of the instances per classes
print('Distribution of the instances per classes')
for col in hotel_reservations.columns:
    if col != 'Booking_ID':
        print(hotel_reservations[col].value_counts())
```

c) Summarize each of the attributes:
- Numerical: mean, standard deviation, min, max, quartiles
- Categorical:  number of distinct values, mode

```python
for col in hotel_reservations.columns:
    if hotel_reservations[col].dtype in ['int64', 'float64']:
        print(f'Statistics for numerical attribute <{col}>:')
        print(f'Mean: {hotel_reservations[col].mean()}')
        print(f'Standard Deviation: {hotel_reservations[col].std()}')
        print(f'Minimum: {hotel_reservations[col].min()}')
        print(f'Maximum: {hotel_reservations[col].max()}')
        print(f'25th Percentile: {hotel_reservations[col].quantile(0.25)}')
        print(f'50th Percentile (Median): {hotel_reservations[col].median()}')
        print(f'75th Percentile: {hotel_reservations[col].quantile(0.75)}')
        print()
    elif hotel_reservations[col].dtype in ['object', 'category']:
        print(f'Statistics for categorical attribute <{col}>:')
        print(f'Number of distinct values: {hotel_reservations[col].nunique()}')
        print(f'Mode: {hotel_reservations[col].mode()[0]}')
        print()
```
